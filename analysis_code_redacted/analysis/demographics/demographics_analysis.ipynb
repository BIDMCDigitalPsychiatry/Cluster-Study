```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/demographics.csv').drop(columns = ['Unnamed: 0'])
```

```python
cols = ['Diagnosis', 'Gender', 'Race']
```

```python
df['Diagnosis'] = df['Diagnosis'].replace('Major Depressive Disorder', 'MDD')
df['Gender'] = df['Gender'].replace('Male', 'M')
df['Gender'] = df['Gender'].replace('Female', 'F')
df['Race'] = df['Race'].replace('White & Asian', 'Asian/White')
df['LAMP ID'] = df['LAMP ID'].replace('Unknown', None)
df = df[cols].dropna().reset_index().drop(columns = ['index'])
```

```python
gender_df = df['Gender'].value_counts().reset_index()
gender_df['pct'] = (gender_df['Gender'] / gender_df['Gender'].sum()).apply(lambda x: np.round(x, 2))
diag_df = df['Diagnosis'].value_counts().reset_index()
diag_df['pct'] = (diag_df['Diagnosis'] / diag_df['Diagnosis'].sum()).apply(lambda x: np.round(x, 2))
race_df = df['Race'].value_counts().reset_index()
race_df['pct'] = (race_df['Race'] / race_df['Race'].sum()).apply(lambda x: np.round(x, 2))
```
